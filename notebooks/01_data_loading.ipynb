```python
import gzip

series_matrix_path = "../data/raw/GSE33000_series_matrix.txt"

sample_ids = []
disease_status = []

with open(series_matrix_path, "r") as f:
    for line in f:
        # Extract sample IDs
        if line.startswith("!Sample_geo_accession"):
            sample_ids = line.strip().split("\t")[1:]
            sample_ids = [s.replace('"', '') for s in sample_ids]

        # Extract disease status
        if line.startswith("!Sample_characteristics_ch2"):
            if "disease status" in line.lower():
                disease_status = line.strip().split("\t")[1:]
                disease_status = [
                    "AD" if "alzheimer" in s.lower() else "Control"
                    for s in disease_status
                ]
```

```python
print(len(sample_ids), len(disease_status))
sample_ids[:5], disease_status[:5]
```

```python
labels_df = pd.DataFrame({
    "sample_id": sample_ids,
    "status": disease_status
})

labels_df.to_csv("../data/processed/sample_labels.csv", index=False)
```
